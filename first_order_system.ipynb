```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import ode
```

System of ODEs:

$$\frac{dx}{dt} = F(x, y)$$

$$\frac{dy}{dt} = G(x, y)$$

With initial conditions $(x(0), y(0)) = (x_0, y_0)$

```python
k = 0.9
l = -0.9
alpha = -0.2
beta = -0.2
g = 0.5
h = 0.7

def F(x, y):
    global k
    global alpha
    global g
    return k*y - alpha*x + g

def G(x, y):
    global l
    global beta
    global h
    return l*x - beta*y + h
```

Initial Conditions

```python
# Initial Condition y0 = [x0, y0]
y0 = [2.0, 1]
```

define a solver function containing the `ode.first_order_system()` class and the `solve_system()` method.

```python
def solver_(F, G, y0):
    # Use the parameters to create an instance of the ode.second_order_system() class
    model = ode.first_order_system(F, G, y0)

    # Solve the equation
    return model.solve_system(), np.arange(model.t_int[0], model.t_int[1], model.dt)
```

Plug in the parameters that are the ODEs coefficients and initial conditions.

```python
(x, y), t = solver_(lambda x, y: F(x, y), lambda x, y: G(x, y), y0) 
```

### Phase Plot:

```python
# Plot 0
plt.figure(0)
plt.plot(x, y, color='green')
plt.title('Phase plot of the model', loc='center')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

### Plots of $x$ and $y$:

```python
# Plot 1
plt.figure(1)
plt.plot(t, x)
plt.plot(t, y)
plt.title('ODE plot', loc='center')
plt.xlabel('$t$')
plt.ylabel('$x$ or $y$')
plt.legend(['$x$', 'y'])
plt.show()
```

### Equilibrium Point

```python
model_ = ode.first_order_system(lambda x, y: F(x, y), lambda x, y: G(x, y), y0)

equilib = model_.equilibrium()

print(f'Equilibrium point is {equilib}')
```

### Plot of $x$ and $y$ tending to equilibrium

$x$ with $t$ plot.

```python
# Plot 2
plt.figure(2)
plt.plot(t, x)
plt.plot(t, [equilib[0] for _ in range(len(t))], color='black')
plt.title('ODE plot', loc='center')
plt.xlabel('$t$')
plt.ylabel('$x$ or $y$')
plt.legend(['$x$', 'y'])
plt.show()
```

$y$ with $t$ plot.

```python
# Plot 3
plt.figure(3)
plt.plot(t, y)
plt.plot(t, [equilib[0] for _ in range(len(t))], color='black')
plt.title('ODE plot', loc='center')
plt.xlabel('$t$')
plt.ylabel('$x$ or $y$')
plt.legend(['$x$', 'y'])
plt.show()
```
