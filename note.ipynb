# Second Order Ordinary Differential Equation Solver Examples

```python
import numpy as np
import matplotlib.pyplot as plt
import ode
```

Second Order ODE:

$$\frac{d^2y}{dt^2} + a(t)\frac{dy}{dt} + b(t)y = c(t)$$

The objective is to solve the above equation with the initial conditions; $y(0) = y_0$ and $y^\prime(0) = y^\prime_0$.

##### Coefficents

```python
# define the coefficient functions
def a(t):
    return float(0)

def b(t):
    return float(1)

def c(t):
    omega = 2*np.pi
    return 0.2*np.cos(omega*t)
```

##### Initial Conditions

```python
# State the intial conditions y0 = [y_0, y'_0]
y0 = [0.5, 0.2]
```

Create an instance of the `ode.second_order_ode()` class with the above parameters (coefficients and initial conditions). Then use the `solve_ode()` method to get a solution. Solutions consist of $y(t)$ and its derivative $y^\prime(t)$.

```python
# Use the parameters to create an instance of the ode.second_order_ode() class
model = ode.second_order_ode(lambda t: a(t), lambda t: b(t), lambda t: c(t), y0)

# Set a time interval
t = np.arange(model.t_int[0], model.t_int[1], model.dt)

# Solve the equation
y, yprime = model.solve_ode()
```

Plotting the solution solution $y$ and its derivative $y^\prime$ over time $t$.

```python
# Plot 0
plt.figure(0)
plt.plot(t, y)
plt.plot(t, yprime)
plt.title('ODE plot', loc='center')
plt.xlabel('$t$')
plt.ylabel('$y$ / $y^\prime$')
plt.legend(['$y$', 'yprime'])
plt.show()
```

Plotting the phase plot of $y$ and $y^\prime$.

```python
# Plot 1
plt.figure(1)
plt.plot(y, yprime, color='red')
plt.title('Phase plot of the model', loc='center')
plt.xlabel('$y$')
plt.ylabel('$y^\prime$')
plt.show()
```
